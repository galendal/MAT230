```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets, Layout
from scipy.integrate import solve_ivp
```

## Exercise i)

$$
\ddot{u}+u =\epsilon (\dot{u}+u), \ u(0)=1, \ \dot{u}(0)   =0 
$$

Expansion: 
$$
u(t)=\frac{2}{\sqrt{4+3\epsilon t}} \cos\left(t(1+\frac{\epsilon}{2})\right) + O(\epsilon)
$$



```python
def rhs_1(t, y, eps=0.1):
    u, dudt = y
    d2xdt2 = -u - eps*(dudt**3 + u)
    return [dudt, d2xdt2]

def asymptotic_1(t, eps=0.1):
    # This is a placeholder for the asymptotic solution of the Van der Pol oscillator
    # The actual asymptotic solution can be derived using perturbation methods
    return 2/np.sqrt(4+3*eps*t)*np.cos(t*(1+0.5*eps)) # Replace with actual asymptotic solution if known

eps=0.1
t_max=80
solver = solve_ivp(rhs_1, [0, t_max], [1, 0], args=(eps,), t_eval=np.linspace(0, t_max, 1000))
t = solver.t
x = solver.y[0] 

fig, axs = plt.subplots(2,1,figsize=(12, 5))
ax=axs[0]
ax.plot(t, x, label=f'Numerical solution (eps={eps})', linewidth=2)

ax.plot(t, asymptotic_1(t, eps), label='Asymptotic Solution', linestyle='-.')
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title('Van der Pol Oscillator Solution')
ax.grid(True)
ax.legend()

ax=axs[1]
ax.plot(t, x-asymptotic_1(t, eps), label=f'asymptotic expansion (μ={eps})', linewidth=2)
ax.set_xlabel('t')
ax.set_ylabel('x(t) - Asymptotic Solution')
ax.set_title('Difference between Numerical and Asymptotic Solutions')
ax.grid(True)
ax.legend()
plt.show()
```

## Exercise iii)

$$
{\displaystyle \frac{d^{2}u}{dt^{2}}+u=\epsilon u\frac{du}{dt}\left(u+\frac{du}{dt}\right),\ u(0)=4,\ \frac{du}{dt}(0)=0}
$$

Expansion:
$$
u(t)=\frac{2}{\sqrt{1/4-\epsilon t}}\cos(t+\frac{1}{2} ln(\frac{1}{4}-\epsilon t)+ln(2)) + O(\epsilon)
$$

```python
def rhs_3(t, y, eps=0.1):
    u, dudt = y
    d2xdt2 = -u + eps*u*dudt*(u+dudt)
    return [dudt, d2xdt2]

def asymptotic_3(t, a=1, eps=0.1):
    # This is a placeholder for the asymptotic solution of the Van der Pol oscillator
    # The actual asymptotic solution can be derived using perturbation methods
    return 2*np.cos(t+0.5*np.log(0.25+eps*t)+np.log(2))/np.sqrt(0.25+eps*t) # Replace with actual asymptotic solution if known

eps=0.1
t_max=2

solver = solve_ivp(rhs_3, [0, t_max], [4, 0], args=(eps,), t_eval=np.linspace(0, t_max, 1000))
t = solver.t
x = solver.y[0] 

fig, axs = plt.subplots(2,1,figsize=(12, 5))
ax=axs[0]
ax.plot(t, x, label=f'Numerical solution (eps={eps})', linewidth=2)

ax.plot(t, asymptotic_3(t, a, eps), label='Asymptotic Solution', linestyle='-.')
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title('Van der Pol Oscillator Solution')
ax.grid(True)
ax.legend()

ax=axs[1]
ax.plot(t, x-asymptotic_3(t, a, eps), label=f'asymptotic expansion (μ={eps})', linewidth=2)
ax.set_xlabel('t')
ax.set_ylabel('x(t) - Asymptotic Solution')
ax.set_title('Difference between Numerical and Asymptotic Solutions')
ax.grid(True)
ax.legend()
plt.show()
```

## Exercise iv)

$$

## Exercis v)
$$
{\displaystyle \frac{d^{2}u}{dt^{2}}+u=\epsilon\left(u^{2}\frac{du}{dt}+u\right),\ u(0)=0,\ \frac{du}{dt}(0)=1}

$$

Expansion:
$$
u(t)=\frac{2}{\sqrt{4-\epsilon t}} \sin(t(1-\frac{\epsilon}{2}))
$$



```python
def rhs_5(t, y, eps=0.1):
    u, dudt = y
    d2xdt2 = -u + eps*(u**2*dudt+u)
    return [dudt, d2xdt2]

def asymptotic_5(t, a=1, eps=0.1):
    # This is a placeholder for the asymptotic solution of the Van der Pol oscillator
    # The actual asymptotic solution can be derived using perturbation methods
    out=2*np.sin(t*(1-eps/2)/np.sqrt(4-eps*t))
    return out # Replace with actual asymptotic solution if known

eps=0.1
t_max=5

solver = solve_ivp(rhs_5, [0, t_max], [0, 1], args=(eps,), t_eval=np.linspace(0, t_max, 1000))
t = solver.t
x = solver.y[0] 

fig, axs = plt.subplots(2,1,figsize=(12, 5))
ax=axs[0]
ax.plot(t, x, label=f'Numerical solution (eps={eps})', linewidth=2)

ax.plot(t, asymptotic_5(t, a, eps), label='Asymptotic Solution', linestyle='-.')
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title('Van der Pol Oscillator Solution')
ax.grid(True)
ax.legend()

ax=axs[1]
ax.plot(t, x-asymptotic_5(t, a, eps), label=f'asymptotic expansion (μ={eps})', linewidth=2)
ax.set_xlabel('t')
ax.set_ylabel('x(t) - Asymptotic Solution')
ax.set_title('Difference between Numerical and Asymptotic Solutions')
ax.grid(True)
ax.legend()
plt.show()
```
