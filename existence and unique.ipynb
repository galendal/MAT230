# Well posed problems and Existence and Uniqueness Theorem

A problem is said to be **well posed** (in the sense of Hadamard) if it satisfies the following three conditions:
1. A solution exists.
2. The solution is unique.
3. The solution's behavior changes continuously with the initial conditions (stability).

If any of these conditions fail, the problem is called **ill posed**.

**Existence and Uniqueness Theorem for Ordinary Differential Equations**

Let $f(t, y)$ be a function continuous in a region containing the point $(t_0, y_0)$. Consider the initial value problem:
$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
$$
for systems of equations, the theorem extends to vector-valued functions $\mathbf{y}(t)$ and $\mathbf{f}(t, \mathbf{y})$, with similar conditions on continuity and partial derivatives ensuring existence and uniqueness of solutions.

If $f(t, y)$ satisfies a **Lipschitz condition** with respect to $y$ in a region containing $(t_0, y_0)$—that is, there exists a constant $L$ such that
$$
|f(t, y_1) - f(t, y_2)| \leq L |y_1 - y_2|
$$
for all $y_1, y_2$ in the region—then the existence and uniqueness of the solution is guaranteed.



**Summary:**  
- *Existence*: A solution exists near $x_0$.
- *Uniqueness*: The solution is unique in that interval.

**Theorem:**
If $f(t, y)$ and its partial derivative with respect to $y$, $\frac{\partial f}{\partial y}$, are continuous in some rectangle containing $(t_0, y_0)$, then there exists an interval $I$ around $t_0$ in which there is a unique function $y(t)$ that solves the initial value problem.



## Examples: Lipschitz Condition and Derivatives

### 1. Lipschitz Condition Example

Consider $f(t, y) = 2y$.  
For any $y_1, y_2$:
$$
|f(t, y_1) - f(t, y_2)| = |2y_1 - 2y_2| = 2|y_1 - y_2|
$$
So $f$ is Lipschitz continuous in $y$ with Lipschitz constant $L = 2$.

---

### 2. Non-Lipschitz Example

Consider $f(t, y) = y^{1/3}$.  
Compute:
$$
|f(t, y_1) - f(t, y_2)| = |y_1^{1/3} - y_2^{1/3}|
$$
As $y_1, y_2 \to 0$, the difference can become arbitrarily large compared to $|y_1 - y_2|$, so $f$ is **not** Lipschitz at $y = 0$.

---

### 3. Discontinuous Derivative Example

Let $f(t, y) = |y|$.  
The derivative with respect to $y$ is:
$\frac{\partial f}{\partial y} = \frac{d}{dy}|y| = 
\begin{cases}
1 & \text{if } y > 0 \\
-1 & \text{if } y < 0 \\
\text{undefined} & \text{if } y = 0
\end{cases}$

For $f(t, y) = |y|$,  
$$
|f(t, y_1) - f(t, y_2)| = ||y_1| - |y_2|| \leq |y_1 - y_2|
$$
So $f$ is Lipschitz continuous in $y$ with Lipschitz constant $L = 1$.  
Even though the derivative is not continuous at $y = 0$, the Lipschitz condition is sufficient for existence and uniqueness.


## Picard iterations
Given the initial value problem:
$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
$$

We can rewrite it as an **integral equation**:
$$
y(t) = y_0 + \int_{t_0}^{t} f(s, y(s))\, ds
$$

**Picard Iterations** are a sequence of approximations defined recursively:
- Start with $y_0(t) = y_0$
- For $n \geq 0$,
    $$
    y_{n+1}(t) = y_0 + \int_{t_0}^{t} f(s, y_n(s))\, ds
    $$

This iterative process converges to the unique solution under suitable conditions (e.g., $f$ is Lipschitz continuous in $y$).
