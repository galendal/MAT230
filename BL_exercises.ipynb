```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

# Exercise 1



## i)

$$
\epsilon f''(x) + f'(x) + \frac{f}{x+1}=2, \  f(0)=0, \ f(1)=3
$$
with expansion
$$
f_0^{\tiny comp}(x) = \frac{x^2+2x+3}{x+1}-3e^{-x/\epsilon}
$$

```python

epsilon=0.01 # small parameter

def f_0_1i(x,epsilon): # Boundary layer expansion
    return (x**2 +2*x+3)/(x+1)-3*np.exp(-x/epsilon)

def rhs_1i(x, z): # Define the right-hand side of the ODE
    return np.vstack((z[1], (2-z[1] - z[0] / (x + 1)) / epsilon))

# Define the boundary conditions
def bc_1i(ya, yb):
    return np.array([ya[0], yb[0] - 3])

# Initial guess for the solution
x_guess = np.linspace(0, 1, 10) # Initial guess for x
y_guess = np.zeros((2, x_guess.size)) # Initial guess for y

sol_bvp = solve_bvp(rhs_1i, bc_1i, x_guess, y_guess) # Solve the BVP

# Plot the solution
plt.figure(figsize=(10, 3))
plt.plot(sol_bvp.x, sol_bvp.y[0], 'bx-',label='solve_bvp solution')
plt.plot(sol_bvp.x, f_0_1i(sol_bvp.x,epsilon), 'g-',label='Boundary layer expansion')
plt.legend()
ax2 = plt.twinx()
sol_diff=sol_bvp.y[0] - f_0_1i(sol_bvp.x, epsilon)
ax2.plot(sol_bvp.x, sol_diff, 'r-',label='Difference')
plt.legend()
plt.show()
```

## ii)

$$
\epsilon f''(x) - f'(x) + \frac{f}{x+1}=2, \  f(0)=0, \ f(1)=3
$$
with expansion
$$
f_0^{\tiny comp}(x) = -2(1+x)\ln(1+x) + (3+4\ln(2))e^{-(1-x)/\epsilon}
$$

```python

epsilon=0.01 # small parameter

def f_0_1ii(x,epsilon): # Boundary layer expansion
    return -2*(1+x)*np.log(1+x)+(3+4*np.log(2))*np.exp(-(1-x)/epsilon)

def rhs_1ii(x, z): # Define the right-hand side of the ODE
    return np.vstack((z[1], (2+z[1] - z[0] / (x + 1)) / epsilon))

# Define the boundary conditions
def bc_1ii(ya, yb):
    return np.array([ya[0], yb[0] - 3])

# Initial guess for the solution
x_guess = np.linspace(0, 1, 10) # Initial guess for x
y_guess = np.zeros((2, x_guess.size)) # Initial guess for y

sol_bvp = solve_bvp(rhs_1ii, bc_1ii, x_guess, y_guess) # Solve the BVP

# Plot the solution
plt.figure(figsize=(10, 3))
plt.plot(sol_bvp.x, sol_bvp.y[0], 'bx-',label='solve_bvp solution')
plt.plot(sol_bvp.x, f_0_1ii(sol_bvp.x,epsilon), 'g-',label='Boundary layer expansion')
plt.legend()
ax2 = plt.twinx()
sol_diff=sol_bvp.y[0] - f_0_1ii(sol_bvp.x, epsilon)
ax2.plot(sol_bvp.x, sol_diff, 'r-',label='Difference')
plt.legend()
plt.show()
```

## iii)

$$
\epsilon f''(x) +x f'(x) + f=0, \  f(2)=0, \ f(4)=1
$$
with expansion
$$
f_0^{\tiny comp}(x) = \frac{4}{x}-2e^{-2(x-2)/\epsilon}
$$

```python

epsilon=0.1 # small parameter

def f_0_1iii(x,epsilon): # Boundary layer expansion
    return 4/x-2*np.exp(-2*(x-2)/epsilon)

def rhs_1iii(x, z): # Define the right-hand side of the ODE
    return np.vstack((z[1], (-x*z[1] - z[0]) / epsilon))

# Define the boundary conditions
def bc_1iii(ya, yb):
    return np.array([ya[0], yb[0] - 1])

# Initial guess for the solution
x_guess = np.linspace(2, 4, 10) # Initial guess for x
y_guess = np.zeros((2, x_guess.size)) # Initial guess for y

sol_bvp = solve_bvp(rhs_1iii, bc_1iii, x_guess, y_guess) # Solve the BVP

# Plot the solution
plt.figure(figsize=(10, 3))
plt.plot(sol_bvp.x, sol_bvp.y[0], 'bx-',label='solve_bvp solution')
plt.plot(sol_bvp.x, f_0_1iii(sol_bvp.x,epsilon), 'g-',label='Boundary layer expansion')
plt.legend()
ax2 = plt.twinx()
sol_diff=sol_bvp.y[0] - f_0_1iii(sol_bvp.x, epsilon)
ax2.plot(sol_bvp.x, sol_diff, 'r-',label='Difference')
plt.legend()
plt.show()
```

## iv)

$$
\epsilon f''(x) - x f'(x) + f=0, \  f(2)=0, \ f(4)=1
$$

$$
f_0^{\tiny comp}(x) = e^{-4(4-x)/\epsilon}
$$

```python

epsilon=0.1 # small parameter

def f_0_1iv(x,epsilon): # Boundary layer expansion
    return np.exp(-4*(4-x)/epsilon)

def rhs_1iv(x, z): # Define the right-hand side of the ODE
    return np.vstack((z[1], (+x*z[1] - z[0]) / epsilon))

# Define the boundary conditions
def bc_1iv(ya, yb):
    return np.array([ya[0], yb[0] - 1])

# Initial guess for the solution
x_guess = np.linspace(2, 4, 10) # Initial guess for x
y_guess = np.zeros((2, x_guess.size)) # Initial guess for y

sol_bvp = solve_bvp(rhs_1iv, bc_1iv, x_guess, y_guess) # Solve the BVP

# Plot the solution
plt.figure(figsize=(10, 3))
plt.plot(sol_bvp.x, sol_bvp.y[0], 'bx-',label='solve_bvp solution')
plt.plot(sol_bvp.x, f_0_1iv(sol_bvp.x,epsilon), 'g-',label='Boundary layer expansion')
plt.legend()
ax2 = plt.twinx()
sol_diff=sol_bvp.y[0] - f_0_1iv(sol_bvp.x, epsilon)
ax2.plot(sol_bvp.x, sol_diff, 'r-',label='Difference')
plt.legend()
plt.show()
```

## v)

# Exercise 2

$$
\epsilon f''(x) + x f'(x) + x f=0, \  f(-1)=e, \ f(1)=2e^{-1}
$$
with expansion
$$
f_0^{\tiny comp}(x) =-\frac{4}{x} -2 e^{2(2+x)/\epsilon}
$$

```python

epsilon=0.1 # small parameter

from scipy.special import erf

def f_0_2(x,epsilon): # Boundary layer expansion
    return (0.5*erf(0.5*x/np.sqrt(epsilon))+1.5)*np.exp(-x)

def rhs_2(x, z): # Define the right-hand side of the ODE
    return np.vstack((z[1], (-x*z[1] - x*z[0]) / epsilon))

# Define the boundary conditions
def bc_2(ya, yb):
    return np.array([ya[0]-np.exp(1), yb[0]-2*np.exp(-1) ])

# Initial guess for the solution
x_guess = np.linspace(-1, 1, 10) # Initial guess for x
y_guess = np.zeros((2, x_guess.size)) # Initial guess for y

sol_bvp = solve_bvp(rhs_2, bc_2, x_guess, y_guess) # Solve the BVP

# Plot the solution
plt.figure(figsize=(10, 3))
plt.plot(sol_bvp.x, sol_bvp.y[0], 'bx-',label='solve_bvp solution')
plt.plot(sol_bvp.x, f_0_2(sol_bvp.x,epsilon), 'g-',label='Boundary layer expansion')
plt.legend()
ax2 = plt.twinx()
sol_diff=sol_bvp.y[0] - f_0_2(sol_bvp.x, epsilon)
ax2.plot(sol_bvp.x, sol_diff, 'r-',label='Difference')
plt.legend()
plt.show()
```
